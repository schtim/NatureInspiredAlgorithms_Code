```python
def plot_solution(objects, optimal_solution, container_vol, container_weight,number_generations, xlim, ylim, pop_size, mutation_prob, crossover_prob, sort_in_recom_and_mutation = True):
    av_number = 10
    amount_obj = len(objects)

    average_vals = np.zeros((av_number, number_generations))
    best_vals = np.zeros((av_number, number_generations))
    runtimes = np.zeros(av_number)
    all_time_bests = np.zeros(av_number)

    fig = [None] * 2
    ax = [None] * 2

    for index in np.arange(av_number):
        GA = GeneticAlgorithm(objects,  container_vol, container_weight,  number_generations, population_size = pop_size, mutation_probability= mutation_prob, crossover_probability= crossover_prob, sort_in_recom_and_mutation = sort_in_recom_and_mutation)
        _,_, all_time_bests[index], average_vals[index], best_vals[index],_,runtimes[index],_ = GA.run()

    # calc average
    average_vals = np.average(average_vals, 0)
    best_vals = np.average(best_vals, 0)
    all_time_best = int(np.min(all_time_bests))
    runtime = np.average(runtimes)

    # Plot
    x_vals = np.arange(number_generations)
    # Punkte visualisieren
    fig[0], ax[0] = plt.subplots(1,figsize=(16,7))
    fig[1], ax[1] = plt.subplots(1,figsize=(16,7))
    fig[0].tight_layout(h_pad=6)
    fig[1].tight_layout(h_pad=6)
    ax[0].set_xlim(xlim)
    ax[0].set_ylim(ylim)
    ax[1].set_xlim(xlim)
    ax[1].set_ylim(ylim)
    ax[0].set_xlabel('Generation/Iteration')
    ax[1].set_xlabel('Generation/Iteration')
    ax[0].set_ylabel("Anzahl der Container")
    ax[1].set_ylabel("Anzahl der Container")
    ax[0].set_title(f'Durchschnittliche Anzahl der Container für {amount_obj} Objekte.)')
    ax[0].plot(x_vals, average_vals, color="blue", label = f'Average Number of Bins in Population. Runtime {runtime}')
    ax[0].plot(x_vals, [optimal_solution]*len(x_vals), color = "black",linestyle = 'dashed', label = f'Optimal Solution ({optimal_solution} Container)')
    ax[1].plot(x_vals, [optimal_solution]*len(x_vals), color = "black",linestyle = 'dashed', label = f'Optimal Solution ({optimal_solution} Container)')
    ax[0].legend()
    print(f'Best Value: {all_time_best} in {np.count_nonzero(all_time_bests == all_time_best)/av_number*100}% of the runs.')
    print(f'Best Value: {all_time_best+1} in {np.count_nonzero(all_time_bests == all_time_best+1)/av_number*100}% of the runs.')
    print(f'Best Value: {all_time_best+2} in {np.count_nonzero(all_time_bests == all_time_best+2)/av_number*100}% of the runs.')

    ax[1].set_title(f'Beste Anzahl der Container für {amount_obj} Objekte.')
    ax[1].plot(x_vals, best_vals, color="blue", label = 'Beste Anzahl der Container in der Population.')
    ax[1].legend()
    plt.show()
```
